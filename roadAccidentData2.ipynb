```python
accidentSeverity = roadAccidentData['Accident_Severity'].value_counts().reset_index()
accidentSeverityDict = accidentSeverity.set_index('index')['Accident_Severity'].to_dict()

accidentSeverity
```

```python
accidentSeverityPerDay = roadAccidentData[['Accident_Severity','Day_of_Week']]
accidentSeverityPerDayDict = dict(accidentSeverityPerDay.value_counts().items())

for key, value in accidentSeverityPerDayDict.items():
    print(f'{key}: {value}')
```

```python
# Transforme o dicionário em um DataFrame
df = pd.DataFrame(list(accidentSeverityPerDayDict.items()), columns=['Dia', 'Contagem'])

# Desmembre a coluna de Dia em duas colunas separadas
df[['Dia', 'Grau']] = pd.DataFrame(df['Dia'].tolist(), index=df.index)

# Pivot a tabela para ter os graus como colunas
df_pivot = df.pivot(index='Dia', columns='Grau', values='Contagem')

# Plote um gráfico de barras empilhadas
df_pivot.plot(kind='bar', stacked=False, rot=0)
plt.xlabel('Dia da Semana')
plt.ylabel('Contagem')
plt.title('Contagem de Incidentes por Dia e Grau de Seriedade')
plt.legend(title='Grau de Seriedade')
plt.show()
```

```python
df = pd.DataFrame(list(accidentSeverityPerDayDict.items()), columns=['Dia', 'Contagem'])
df[['Dia', 'Grau']] = pd.DataFrame(df['Dia'].tolist(), index=df.index)
df_pivot = df.pivot(index='Grau', columns='Dia', values='Contagem')
df_pivot.plot(kind='bar', stacked=True, rot=0)
plt.xlabel('Grau de Seriedade')
plt.ylabel('Contagem')
plt.title('Contagem de Incidentes por Grau de Seriedade e Dia da Semana')
plt.legend(title='Dia da Semana')
plt.show()
```
