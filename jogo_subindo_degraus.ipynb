**Imagine que você está subindo as escadas de um prédio com um amigo/amiga seu e que vocês decidem jogar um jogo pra ver se em 100 rolagens de dado, você  conseguirá subir até o degrau 60. Existem as seguintes regras:**



Se o dado der o número 01 ou 02, você desce um degrau.
Caso dê 2, 3 ou 4 vocês podem subir um degrau. E, dando 6, vocês podem jogar novamente o dado para descobrir quantos degraus vocês irão subir, ou seja, qualquer que seja o número sorteado imaginando ele como X, vocês irão subir X degraus. Você não pode descer abaixo do degrau 0. E por fim, existe uma pequena chance de você cair até láaaa embaixo, uma chance de 0,5% diremos.

```python
# Import numpy and set seed
import numpy as np
np.random.seed(123)

# Use randint() to simulate a dice
print(np.random.randint(1,7))

# Use randint() again
print(np.random.randint(1,7))
```

Vamos supor que você esteja no degrau 50, por exemplo. Como iríamos fazer?

```python
# Starting step
step = 50

# Roll the dice
dice = np.random.randint(1,7)

# Finish the control construct
if dice <= 2 :
    step = step - 1
elif dice <= 5 :
    step = step + 1
else :
    step = step + np.random.randint(1,7)

# Print out dice and step
print(dice)
print(step)
```

Aqui vemos que você estava no degrau 50 e que você jogou o dado, obtendo o número 4. Como previamente dito nas regras, caso o número no dado seja 3, 4 ou 5 você sobe apenas um degrau. Portanto, estaremos no degrau número 51.

Vamos tentar algo um pouco diferente? Anteriormente nós definimos o próximo degrau com base no último. Isso é conhecido como passo aleatório! Vamos tentar agora uma "jornada" aleatória, onde usamos diversos elementos anteriores para definir o próximo passo.

```python
# Initialize random_walk
random_walk = [0]

# Complete the ___
for x in range(100) :
    # Set step: last element in random_walk
    step = random_walk[-1]

    # Roll the dice
    dice = np.random.randint(1,7)

    # Determine next step
    if dice <= 2:
        step = step - 1
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    # append next_step to random_walk
    random_walk.append(step)

# Print random_walk
print(random_walk)
```

As coisas estão começando a ter efeito, porém, algo se esqueceu! Nas regras do jogo nós definimos que você não pode passar para baixo do degrau 0, portanto, não tem como haver números negativos representando degraus, como vemos anteriormente.

**Uma forma de nós resolvermos isso é utilizando a função max(). Se você passar para a função max() dois argumentos, o maior argumento se retorna. Por exemplo, caso quisessemos que uma variável X nunca fosse para abaixo do número 5, poderiamos usar:**

```python
x = max(5, x - 1)
```

Assim, definimos que o menor valor deverá ser 5 e que o utilizado deve ser o maior valor em relação a 5. No nosso caso, em que queremos que o menor degrau seja 0, utilizamos:

```python
# Inicialize o caminho aleatório
random_walk = [0]

for x in range(100) :
    step = random_walk[-1]
    dice = np.random.randint(1,7)

    if dice <= 2:
        # Substitua abaixo para garantir que não vá para baixo do degrau 0.
        step = max(0, step - 1)
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    random_walk.append(step)

print(random_walk)
```

Pronto! Agora nós vemos o resultado de jogar o dado 100 vezes e garantimos que não existam números negativos. Mas parando para pensar, apenas visualizar rodando o código não é muito intuitivo. Que tal visualizarmos isso aqui de forma melhor, talvez usando o matplotlib?

```python
# Initialization
random_walk = [0]

for x in range(100) :
    step = random_walk[-1]
    dice = np.random.randint(1,7)

    if dice <= 2:
        step = max(0, step - 1)
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    random_walk.append(step)

# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Plot random_walk
plt.plot(random_walk)

# Show the plot
plt.show()
```

Agoooora sim! Muito melhor de se enxergar como a progressão se deu enquanto você jogava o dado 100 vezes!

Mas vamos pensar: Simular apenas 1 vez essa progressão de degraus, não me diz nada sobre a minha chance de ganhar a aposta! Para saber qual a chance de eu alcançar o degrau número 50, preciso fazer essa mesma jornada aleatória de ficar jogando o dado um mooonte de vezes! Só assim, saberei a probabilidade de eu ganhar a aposta. Porém, por fins didáticos, vamos realizar essa jornada aleatória 5 vezes.

```python
# Initialize all_walks (don't change this line)
all_walks = []

# Simulate random walk five times
for i in range(5) :

    # Code from before
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)

    # Append random_walk to all_walks
    all_walks.append(random_walk)

# Print all_walks
print(all_walks)
```

O elemento all_walks, traduzido de forma bruta como "todas as caminhadas" é uma lista contendo outras listas, já que dentro dele cada sub-lista representa uma jornada aleatória única. Se convertermos isso para uma matriz do NumPy, poderemos fazer gráficos diferenciados e interessantes!

```python
# initialize and populate all_walks
all_walks = []
for i in range(5) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)
    all_walks.append(random_walk)

# Convert all_walks to NumPy array: np_aw
np_aw = np.array(all_walks)

# Plot np_aw and show
plt.plot(np_aw)
plt.show()
# Clear the figure
plt.clf()

# Transpose np_aw: np_aw_t
np_aw_t = np.transpose(np_aw)

# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()
```

Acima vemos como diferentes simulações foram e, vimos a importância do recurso transpose(). O recurso transpose() do NumPy é usado para trocar as linhas pelas colunas em uma matriz. Imagine que você tem uma tabela com várias linhas e colunas. A transposição muda a orientação dessa tabela, trocando as linhas pelas colunas.

```python
1  2  3
4  5  6
```

Essa matriz acima, utilizando o transpose() vira:


```python
1  4
2  5
3  6
```

Com esse código nosso, trocar a quantidade de vezes que a jornada aleatória acontece é moleza. Mas, esquecemos de que você é uma pessoa desengonçada e que existe uma chance de 0.5% de você cair até lá embaixo:

```python
# clear the plot so it doesn't get cluttered if you run this many times
plt.clf()

# Simulate random walk 20 times
all_walks = []
for i in range(20) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        # Implement clumsiness
        if np.random.rand() <= 0.005 :
            step = 0

        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))
plt.plot(np_aw_t)
plt.show()
```

Perfeito! Vemos que, mesmo a chance sendo pequena, você continua sendo desengonçado e por consequência você caiu algumas vezes até o degrau inicial.

E agora, focando no principal e deixando todas essas ótimas visualizações de lado, vamos ao que interessa, a pergunta de um milhão de doláres! **Qual é a probabilidade de que você chegará até o degrau 60 no prédio?**


```python
# Simulate random walk 500 times
all_walks = []
for i in range(500) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        if np.random.rand() <= 0.001 :
            step = 0
        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))

# Select last row from np_aw_t: ends
ends = np_aw_t[-1, :]

# Plot histogram of ends, display plot
plt.hist(ends)
plt.show()
```

Simulando 500 vezes, temos aí um histograma demonstrando qual a frequência de cada degrau como o último, mas ainda não a probabilidade.

```python
np.mean(ends>=60)
```

Vemos, agora sim, a probabilidade de você passar pelo degrau 60 sendo de 79.4%!
